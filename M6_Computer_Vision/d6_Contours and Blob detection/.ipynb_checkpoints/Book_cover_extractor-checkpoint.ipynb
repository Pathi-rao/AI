```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
print('setup complete')
```

```python
#helper function
def imshow(img):
    plt.figure(figsize=(10,7))
    plt.imshow(img, cmap='gray')
```

```python

def draw_contours(image):
    img = cv2.imread(image)
    imshow(img)
    
    #convert to grayscale and do some pre-processing
    gray_img = cv2.cvtColor(img.copy(), cv2.COLOR_BGR2GRAY)
    blur_img = cv2.GaussianBlur(gray_img, (3, 3), 0)

    edged = cv2.Canny(blur_img, 10, 250)
    
    contr_copy = img.copy()
    contours, h = cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    x, y, width, height = cv2.boundingRect(edged)
    #The cv2.boundingRect() function of OpenCV is used to draw an approximate rectangle around the 
    #binary image. This function is used mainly to highlight the region of interest after obtaining 
    #contours from an image
    
    print( 'point1:',x,y, 'point2:',x+width,y, 'point3:',x,height+y ,'point4:',width+x, height+y)
    
    cv2.drawContours(contr_copy, contours, -1, (0,255,0), 2)
    imshow(contr_copy)
    
    img = contr_copy.copy()
    #original_coordinates = np.float32([[267,113], [525,500], [250,490], [540,490]])
    original_coordinates = np.float32([[267,113], [525,113], [267,500], [525,500]])
    # Now write the coordinates of the 4 corners of the desired output
    new_coordinates = np.float32([[0,0], [420,0], [0,594], [420,594]])
    # Use the two sets of four points to compute the Perspective Transformation matrix, M  with getPerspectiveTransform  
    M  = cv2.getPerspectiveTransform(original_coordinates, new_coordinates)
    # Apply the perspective transformation matrix 
    warped = cv2.warpPerspective(img, M, (420,594))
    plt.imshow(warped);

draw_contours('img/clean_book.jpg')
```
