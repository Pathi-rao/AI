- If you are interested in preserving the edges, go with median blurring or bilateral filtering. On the   contrary, gaussian blurring is likely to be faster than median blurring. Due to its computational    complexity, bilateral filtering is the slowest of all methods.

- When you look at the matrix values of the gray scale image, we see that it consists of values between 0 and 255. In some cases, we may want this matrix to consist only of values 0 and 255 . The threshold function is used in such cases

```python
image = cv2.imread('self_mnist.jpg')
img = image.copy()

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
resized = cv2.resize(gray_img, (28, 28), interpolation = cv2.INTER_AREA)

plt.imshow(gray_img)

# cv2.imshow('Gray image', gray_img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
#gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
(thresh, output2) = cv2.threshold(resized, 120, 255, cv2.THRESH_BINARY)
#plt.imshow(output2, cmap = plt.get_cmap("gray"))
output2 = cv2.GaussianBlur(output2, (5, 5), 1)
output2 = cv2.Canny(output2, 180, 255)
#resized = cv2.resize(gray_img, (28, 28), interpolation = cv2.INTER_AREA)
#plt.imshow(resized, cmap = plt.get_cmap("gray"))
```
