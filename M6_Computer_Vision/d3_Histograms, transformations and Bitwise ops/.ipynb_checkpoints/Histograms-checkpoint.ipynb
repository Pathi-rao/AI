# Computer Vision Histograms

### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Histograms

```python
# Load some images, calculate and plot their color histograms, one color at the time
# try it with different color spaces BGR, HSV, etc
img = cv2.imread('img/day1.jpg')
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hsv_image = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2HSV)
```

```python
plt.imshow(rgb_image)
```

[Hint](# "cv2.calcHist") 


```python
# Now try to plot all 3 channels at the same time, does the histogram give you an idea of what the image looks like?
colors = ['r','g','b']
for i,color in enumerate(colors):
    hist = cv2.calcHist([img],[i],None,[255],[0,255])
    plt.plot(hist,color = color)
```

```python
plt.imshow(hsv_image)
```

```python

colors = ['r','g','b']
for i, color in enumerate(colors):
    hist = cv2.calcHist([hsv_image],[i],None,[255],[0,255])
    plt.plot(hist,color = color)
```

```python
#function which tells if an image is day or night

def img_estim(img, thrshld):
    image = cv2.imread(img)
    is_light = np.mean(image) > thrshld
    return 'light' if is_light else 'dark'

print(img_estim('img/day1.jpg', 127))
# (127 - 255) denotes light image
# (0 - 127) denotes dark image
```
