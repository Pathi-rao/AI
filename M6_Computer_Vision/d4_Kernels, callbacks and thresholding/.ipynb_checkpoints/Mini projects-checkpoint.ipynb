## Level 1

### Photo art app

1. Create a function that receives an image and converts it to a sketch so it looks similar to this:
![Example 1](img/sketch.png)

````python
#pencil art
def sketech(image):
    img = cv2.imread(image)
    
    cv2.imshow('Original Image', img) 
    k = cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    inverted_gray_img = 255 - gray_img
    
    cv2.imshow('Inverted gray_image', inverted_gray_img) 
    k = cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    blur_img = cv2.GaussianBlur(inverted_gray_img, (21, 21), 0)
    inverted_blur_img = 255 - blur_img
    
    cv2.imshow('Inverted blur image', inverted_blur_img) 
    k = cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    final_img = cv2.divide(gray_img, inverted_blur_img, scale=250)
    
    cv2.imshow('Final_image', final_img) 
    k = cv2.waitKey(0)
    cv2.destroyAllWindows()
    
#     file_name = 'Sketch_art'+ image
#     cv2.imwrite(file_name, final_img)
    
sketech('img/Lenna.png')
````

````python
# filter
image = cv2.imread('img/Lenna.png')

plt.imshow(image, cmap = 'gray')#BGR image

canny_image = cv2.Canny(image, 100, 200)
plt.figure(figsize = (6,6))
plt.imshow(canny_image, cmap = 'gray'); #canned image

final_art = cv2.bitwise_not(canny_image)
plt.figure(figsize = (6,6))
plt.imshow(final_art, cmap = 'gray');
````

### Instagram filters app

1. Create a function that given a photo it applies a black and white filter
1. Create a filter that will change color of the image 

````python
def all_filters(image):
    img = cv2.imread(image)
    cv2.imshow('Original image', img)

    
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    thresh, blk_n_white = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
    
    cv2.imshow('Black white image', blk_n_white)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def filter1(image):
    img = cv2.imread(image)
    filtr = cv2.bitwise_not(img)
    cv2.imshow('negative filter', filtr)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
#black_n_white('img/Lenna.png')
filter1('img/Lenna.png')
````

## Level 2

### Photo art app

1. Change the app so it can now do the same but using your webcam to make it in real time  
**Hint:** you can use ```cv2.VideoCapture(0)```

````python
cap = cv2.VideoCapture(0)

if (cap.isOpened()== False):
    print("Error opening video stream or file")

while(cap.isOpened()):
    
  # Capture frame-by-frame
    ret, frame = cap.read()
    
    if ret == True:
        cv2.imshow('Frame', frame) # Display the original frame
        
        #apply canny filter
        canny_image = cv2.Canny(frame, 150, 250)
        final_art = cv2.bitwise_not(canny_image)
        cv2.imshow('sketch image', final_art)
        
        #apply black and white filter
        gray_img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        thresh, blk_n_white = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
        cv2.imshow('black_n_white', blk_n_white)
        
        #apply negative filter
        filtr = cv2.bitwise_not(frame)
        cv2.imshow('negative filter', filtr)
        
        # Press Q on keyboard to  exit
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
        
  # Break the loop
    else:
        break 
    
# When everything done, release the video capture object
cap.release()
# Closes all the frames
cv2.destroyAllWindows()
    
````

### Instagram filters app

1. Create a function that applies a sepia filter to the image, for a BGR image apply this kernel:
        [0.272, 0.534, 0.131],
        [0.349, 0.686, 0.168],
        [0.393, 0.769, 0.189]
        
Get the BGR value of the pixel.
Calculate tr, tg and tb using the formula

tb = 0.272R + 0.534G + 0.131B
tg = 0.349R + 0.686G + 0.168B
tr = 0.393R + 0.769G + 0.189B

Take the integer value.

Set the new RGB value of the pixel as per the following condition:

If tb > 255 then b = 255 else b = tb
If tg > 255 then g = 255 else g = tg
If tr > 255 then r = 255 else r = tr


Example:

Consider a color pixel with the following values

A = 255

R = 100

G = 150

B = 200

Where A, R, G and B represents the Alpha, Red, Green and Blue value of the pixel.

Remember! ARGB will have an integer value in the range 0 to 255.

So, to convert the color pixel into sepia pixel we have to first calculate tr, tg and tb.

tr = 0.393(100) + 0.769(150) + 0.189(200)

tr = 192.45

tr = 192 (taking integer value)

Similarly,

tg = 0.349(100) + 0.686(150) + 0.168(200) = 171 (taking integer value)

and tb = 0.272(100) + 0.534(150) + 0.131(200) = 133 (taking integer value)
