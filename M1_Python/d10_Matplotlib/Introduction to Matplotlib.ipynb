```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Figures and Axes

```python
fig = plt.figure()
fig.add_subplot(2, 2, 1) # 2 rows, 2 cols, first element
fig.add_subplot(2, 2, 4) # 2 rows, 2 cols, last element
fig.show()
```

```python
fig = plt.figure()
fig.add_subplot(2, 2, 2) # 2 rows, 2 cols, second element
fig.add_subplot(2, 2, 3) # 2 rows, 2 cols, third element
fig.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_xticks([-2, -1, 0, 1, 2])
ax.set_xticklabels(['-std2', '-std1', 'mean', 'std1', 'std2'])
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_xticks([-2, -1, 0, 1, 2])
ax.set_xticklabels(['-std2', '-std1', 'mean', 'std1', 'std2'])
ax.set_yticks([-2, -1, 0, 1, 2])
ax.set_yticklabels(['-y_tick2', '-y_tick1', 'y_tick', 'y_tick1', 'y_tick2'])
```

```python
plt.scatter(
    np.random.normal(size=10), np.random.normal(size=10)
)
```

```python
plt.plot(
    np.arange(-5, 6, 0.01), np.sin(np.arange(-5, 6, 0.01))
)
```

You can constrain the x and y axis values with `xlim` and `ylim` methods:

```python

plt.ylim(-0.5,0.5)
plt.xlim(-3,3)
plt.plot(
    np.arange(-5, 6, 0.01), np.sin(np.arange(-5, 6, 0.01))
)
```

If you take a deeper look at the [matplotlib.pyplot.plot documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) you can see how to tune the various arguments to produce more specific plots:


```python

plt.xlim(-3,3)
plt.plot(
    np.arange(-5, 6, 0.01), np.sin(np.arange(-5, 6, 0.01)), color='r', marker='o',
    linestyle='--', linewidth=3, markersize=1, markerfacecolor="k"
)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.legend(loc="upper right") # no need in this example
plt.title("This is the title")
plt.show()
```
